```python
#FULL REPORT

from nba_api.stats.endpoints import leaguedashptstats
from nba_api.stats.static import teams
import pandas as pd
from datetime import date
from tabulate import tabulate

'''
PLAY TYPES
speed_distance = 'SpeedDistance'
rebounding = 'Rebounding'
possessions = 'Possessions'
catch_shoot = 'CatchShoot'
pull_up_shot = 'PullUpShot'
defense = 'Defense'
drives = 'Drives'
passing = 'Passing'
elbowTouch = 'ElbowTouch'
postTouch = 'PostTouch'
paintTouch = 'PaintTouch'
efficiency = 'Efficiency'
'''

season='2023-24'
play_types= ['Drives','Rebounding', 'SpeedDistance', 'CatchShoot', 'PullUpShot', 'Defense', 'Passing', 'ElbowTouch', 'PostTouch', 'PaintTouch', 'Efficiency']

for play_type in play_types:
    league_dash_pt_stats = leaguedashptstats.LeagueDashPtStats(
        per_mode_simple='PerGame',
        player_or_team='team',
        pt_measure_type=play_type,
        season=season,
        proxy=None,
        headers=None,
        timeout=30,
        get_request=True
    )

    results = league_dash_pt_stats.get_data_frames()[0]

    #remove the TEAM_ID, TEAM_ABBREVIATION, MIN, GP columns
    results = results.drop(columns=['TEAM_ID', 'TEAM_ABBREVIATION'])

    print(tabulate(results, headers='keys', tablefmt='psql'))

    #save to csv with season, play type and date in the file name in a folder named 'stats/tracking/team'
    today = date.today()
    results.to_csv('stats/tracking/team/' + season + '_' + play_type + '_' + str(today) + '.csv', index=False)

```

```python
#DATE FILTERED REPORT

from nba_api.stats.endpoints import leaguedashptstats
from nba_api.stats.static import teams
import pandas as pd
from datetime import date
from tabulate import tabulate

'''
PLAY TYPES
speed_distance = 'SpeedDistance'
rebounding = 'Rebounding'
possessions = 'Possessions'
catch_shoot = 'CatchShoot'
pull_up_shot = 'PullUpShot'
defense = 'Defense'
drives = 'Drives'
passing = 'Passing'
elbowTouch = 'ElbowTouch'
postTouch = 'PostTouch'
paintTouch = 'PaintTouch'
efficiency = 'Efficiency'
'''

#These are my filters
season ='2022-23'
play_type = 'PullUpShot'
datefrom = '1/26/2023'
dateto = '2/27/2023'
location = '' 
outcome = ''

league_dash_pt_stats = leaguedashptstats.LeagueDashPtStats(
        per_mode_simple='PerGame',
        player_or_team='team',
        pt_measure_type=play_type,
        date_from_nullable=datefrom,
        date_to_nullable=dateto,
        location_nullable=location,
        outcome_nullable=outcome,
        season=season,
        proxy=None,
        headers=None,
        timeout=30,
        get_request=True
    )

results = league_dash_pt_stats.get_data_frames()[0]

#remove the TEAM_ID, TEAM_ABBREVIATION, MIN, GP columns
results = results.drop(columns=['TEAM_ID', 'TEAM_ABBREVIATION', 'MIN', 'GP'])

print(tabulate(results, headers='keys', tablefmt='psql'))

#save to csv with the filters I've applied added to the name in a folder named 'stats/tracking/team' make sure to remove slashes from date to and date from
today = date.today()
datefrom = datefrom.replace('/', '')
dateto = dateto.replace('/', '')
results.to_csv('stats/tracking/team/league_' + season + '_' + play_type + '_' + datefrom + '_' + dateto + '_' + location + '_' + outcome + '_' + str(today) + '.csv', index=False)
```
