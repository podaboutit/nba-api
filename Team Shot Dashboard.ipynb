```python
# SHOT DASHBOARD ONE TEAM
from nba_api.stats.endpoints import teamdashptshots
from nba_api.stats.static import teams
import pandas as pd
from datetime import date
from tabulate import tabulate

team = "Charlotte Hornets"
season = "2022-23"

team_dash_pt_shots = teamdashptshots.TeamDashPtShots(
    team_id=teams.find_teams_by_full_name(team)[0]["id"],
    per_mode_simple="PerGame",
    season=season,
    proxy=None,
    headers=None,
    timeout=30,
    get_request=True,
)
```

```python
# SHOT DASHBOARD ALL TEAMS

from nba_api.stats.endpoints import teamdashptshots
from nba_api.stats.static import teams
import pandas as pd
from datetime import date
from tabulate import tabulate

season = "2022-23"

nba_teams = teams.get_teams()
team_ids = [team["id"] for team in nba_teams]

for team in team_ids:
    team_dash_pt_shots = teamdashptshots.TeamDashPtShots(
        team_id=team,
        per_mode_simple="PerGame",
        season=season,
        proxy=None,
        headers=None,
        timeout=30,
        get_request=True,
    )
    results = team_dash_pt_shots.closest_defender_shooting.get_data_frame()
    print(tabulate(results, headers=results.columns, tablefmt="psql"))
```
